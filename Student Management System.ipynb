```python
def show_list(student_list):
    """Show the list of students."""
    print("Here is the Student list:")
    for student in student_list:
        print(student)

def add_student(student_list):
    """Add a new student to the list."""
    new_name = input("Enter the new name to be added in the list: ")
    student_list.append(new_name)
    print("After adding the new name, the list is:", student_list)

def remove_student(student_list, name):
    """Remove a student from the list."""
    if name in student_list:
        student_list.remove(name)
        print(f"{name} removed from the list. Now the list is:", student_list)
    else:
        print(f"{name} is not in the list.")

def search_student(student_list, name):
    """Search for a student in the list."""
    if name in student_list:
        index = student_list.index(name)
        print(f"{name} found at index {index}.")
    else:
        print(f"{name} not found in the list.")

student_list = ['Faheem', 'Divya', 'Khushi', 'Shubham', 'Deepak', 'Mayank']

while True:
    print("\nMenu:")
    print("1. Show Student List")
    print("2. Add Student")
    print("3. Remove Student")
    print("4. Search Student")
    print("5. Exit")

    choice = int(input("\nEnter your choice: "))
    if choice == 1:
        show_list(student_list)
    elif choice == 2:
        add_student(student_list)
    elif choice == 3:
        name_to_remove = input("Enter the name to be deleted: ")
        remove_student(student_list, name_to_remove)
    elif choice == 4:
        name_to_search = input("Enter the name to search for: ")
        search_student(student_list, name_to_search)
    elif choice == 5:
        print("Exiting the program.")
        break
    else:
        print("Enter a valid choice.")
```
